```python
import numpy as np
import pandas as pd 
import math
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
def PointsInCircum(r,n=100):
    return[(math.cos(2*math.pi/n*x)*r+np.random.normal(-30,30),math.sin(2*math.pi/n*x)*r+np.random.normal(-30,30))for x in range(1,n+1)]
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(PointsInCircum(500, 1000))

df = pd.concat([df, pd.DataFrame(PointsInCircum(300, 700))], ignore_index=True)
df = pd.concat([df, pd.DataFrame(PointsInCircum(100, 300))], ignore_index=True)

df = pd.concat([
    df,
    pd.DataFrame(
        [(np.random.randint(-600, 600), np.random.randint(-600, 600)) for _ in range(300)]
    )
], ignore_index=True)
```

```python
df = pd.DataFrame(
    PointsInCircum(500, 1000)
    + PointsInCircum(300, 700)
    + PointsInCircum(100, 300)
    + [(np.random.randint(-600,600), np.random.randint(-600,600)) for _ in range(300)]
)
df
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df[0],df[1],s=15,color="m")
plt.title("dataset",fontsize=18)
plt.xlabel("feature1",fontsize=12)
plt.ylabel("feature2",fontsize=12) 
plt.show

```

```python
from sklearn.cluster import KMeans
```

```python
Kmeans=KMeans(n_clusters=4,random_state=42)
Kmeans.fit(df[[0,1]])
```

```python
df['KMeans_labels'] = Kmeans.labels_
print(df['KMeans_labels'])
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
colors=['purple','red','blue','green']
plt.figure(figsize =(6,6))
plt.scatter(df[0],df[1],c=df['KMeans_labels'],cmap=ListedColormap(colors),s=15)
plt.title('K-Means Clustaring',fontsize=20)
plt.xlabel("Feature 1",fontsize=14)
plt.ylabel("Feature 2",fontsize=14)
plt.show()
```
